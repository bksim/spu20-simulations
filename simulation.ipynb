# Random Walks and Diffusion

The existence of statistical laws in nature may strike one (who hasn't taken SPU 20, for example) as unexpected, perhaps even surpising. After all, things around us seem pretty orderly, and we seem to be able to predict things about the physical world quite well. But the foundation of much (if not all - consider quantum mechanics!) of our understanding of the physical world and the laws which govern it are predicated on our understanding of statistics and the statistical underpinnings of physical phenomena. One interesting physical phenomenon which is explained through statistics is that of diffusion. Consider a particle, located at some initial location (let's say, 0). If at equally spaced intervals, the particle moves left or right some amount with equal probability, what does that look like? [Run the code below to see sample paths of a random walk.]

```python
import numpy as np
import matplotlib.pyplot as plt
import random
%matplotlib inline

##################################################################################################################
timeSteps = 100 ## ALERT! This can be changed to make the simulation last a longer or shorter amount of time.
stepSize = 1    ## ALERT! This can be changed to make the particle move further in each step.
numPaths = 100  ## ALERT! Controls the number of sample paths drawn 
##################################################################################################################

plt.figure()
for p in xrange(numPaths):
    locations = [0]
    for t in xrange(timeSteps):
        locations.append(locations[-1] + np.random.choice([-stepSize, stepSize]))
    plt.plot(range(timeSteps+1), locations, '-')
plt.grid()
plt.xlabel('Time')
plt.ylabel('Location')
plt.title('Path traced out by ' + str(numPaths) + ' randomly walking particle(s) which start at y = 0')
plt.show()
```

The particle appears to undergo a randomly snaking pattern, moving up and down haphazardly. What if we take many of these randomly walking particles and have them all start at 0, then walk for some time and make a plot of where they are? [Run code below.]

```python
##################################################################################################################
particles = 100000     ## ALERT! This controls how many particles are simulated.
timeSteps = 200        ## ALERT! This controls how long the simulation lasts.
stepSize = 1           ## ALERT! This controls how far a particle moves in each step
numSimulations = 1     ## ALERT! This controls how many distributions are plotted.
##################################################################################################################

    
plt.figure()
for s in xrange(numSimulations):
    locations = np.zeros(particles)

    for t in xrange(timeSteps):
        locations = locations + np.random.choice([-stepSize, stepSize], particles)
    plt.hist(locations, 20, alpha=0.3)
    averagePosition = np.mean(locations)
    stdDeviation = np.std(locations)
    
    print "Average position of particles for simulation", s+1, ":", averagePosition
    print "Standard deviation of particles for simulation", s+1, ":", stdDeviation

plt.grid()
plt.show()
```

Recall the meandering path taken by one particle in one dimension earlier. What if the particle now moves randomly in two dimensions? This is similar to what a gas particle does, for example, since all the other gas particles around it are randomly bumping into it and pushing it to move in random directions. This phenomenon is known as diffusion. This explains why, although gas particles are moving very rapidly (over 400 m/s) in the air, they are not actually going that far; if you track one molecule, it might still be in the same room minutes later. In the simulation below, notice how far the particle gets when it randomly moves up, down, left, or right at each time, even for a large number of time steps. Random walks entirely explain diffusion (one can derive the diffusion equation from a random walk minus physical constants, such as diffusion coefficients), and this relationship explains one of the ways that statistics affect physical phenomena [Run code below.]

```python
##################################################################################################################
timeSteps = 1000 ## ALERT! This can be changed to make the simulation last a longer or shorter amount of time.
stepSize = 1    ## ALERT! This can be changed to make the particle move further in each step.
numPaths = 1    ## ALERT! Controls the number of sample paths drawn 
##################################################################################################################

plt.figure()
for p in xrange(numPaths):
    loc_x = [0]
    loc_y = [0]
    for t in xrange(timeSteps):
        choice = np.random.choice(4)
        if choice == 0: #left
            loc_x.append(loc_x[-1] - stepSize)
            loc_y.append(loc_y[-1])
        elif choice == 1: #down
            loc_x.append(loc_x[-1])
            loc_y.append(loc_y[-1] - stepSize)
        elif choice == 2: #right
            loc_x.append(loc_x[-1] + stepSize)
            loc_y.append(loc_y[-1])
        elif choice == 3: #up
            loc_x.append(loc_x[-1])
            loc_y.append(loc_y[-1] + stepSize)
    plt.plot(loc_x, loc_y, '-')
plt.grid()
plt.xlabel('x location')
plt.ylabel('y location')
plt.title('Path traced out over ' + str(timeSteps) + 'steps, by ' + str(numPaths) + ' randomly walking particle(s) which start at (0,0) in 2-D')
plt.show()
```
